```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
#import thư viện
```

Câu 1: Hãy đọc và hiển thị ảnh có tên 1.jpg trong tập dữ liệu trên

```python
img = cv2.imread(r"C:\Users\ldraw\Documents\AIO 2026\Keep_tracking_warm_up_1\Day 10\dataset\1.jpg" )
# hàm dùng đọc thư viện
# r để đọc chỗ "\1" mà không có lỗi
```

```python
print("Origin img")
plt.imshow(img,cmap='gray')
plt.show()
# dùng thư viện matplotlib để hiện thị hình ảnh
```

Câu 2: Chuyển ảnh màu thành xám và hiển thị

```python
img = cv2.cvtColor (img,cv2.COLOR_BGR2GRAY )
# chuyển ảnh thành màu xám
print("Grey img")
plt.imshow(img,cmap='gray')
plt.show()
# dùng thư viện matplotlib để hiển thị hình ảnh
```

Câu 3: Tăng cường độ sáng ảnh màu lên 50 đơn vị, và giảm cường độ sáng xuống 80 đơn vị ( Trên ảnh màu gốc ban đầu ) và hiển thị

* Tăng cường độ sáng ảnh màu lên 50 đơn vị

```python
print("Brightness img")
img1 = cv2.imread(r"C:\Users\ldraw\Documents\AIO 2026\Keep_tracking_warm_up_1\Day 10\dataset\1.jpg" )
img1 = img1.astype(np.float32) + 50
# chuyển đổi ma trận gốc thành dạng 32 bit rồi cộng thêm 50 điểm vào từng pixel
img1 = np.clip(img1,0,255)
# chống gây tràn bit
img1 = img1.astype(np.uint8)
# quay trở lại quy chuẩn số nguyên 8 bit
plt.imshow(img1,cmap='gray')
plt.show()
```

* Giảm cường độ sáng ảnh xuống 80 đơn vị

```python
print("Darkness img")
img2 = cv2.imread(r"C:\Users\ldraw\Documents\AIO 2026\Keep_tracking_warm_up_1\Day 10\dataset\1.jpg" )
img2 =img2.astype(np.float32) - 80
# chuyển đổi ma trận gốc thành dạng 32 bit rồi trừ 80 điểm vào từng pixel
img2 = np.clip(img2,0,255)
# chống gây tràn bit
img2 = img2.astype(np.uint8)
# quay trở lại quy chuẩn số nguyên 8 bit
plt.imshow(img2,cmap='gray')
plt.show()
```
