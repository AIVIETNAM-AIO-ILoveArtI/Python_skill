```python
#Yêu cầu 2
name = "ms Taylor"
#tạo biến name để lưu tên Taylor
hoa_chu_dau = name.title()
#viết hoa chữ cái đầu của chữ trong biến name
viet_hoa = name.upper()
#viết hoa hết tất cả chữ cái của chữ trong biến name
viet_thg = name.lower()
#viết thường hết tất cả chữ cái của chữ trong biến name
print(f"Title case: {hoa_chu_dau}")
print(f"Upper case: {viet_hoa}")
print(f"Lower case: {viet_thg}")
#in kết quả
```
