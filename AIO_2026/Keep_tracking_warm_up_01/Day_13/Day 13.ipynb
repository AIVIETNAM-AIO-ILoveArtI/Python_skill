Bài 2: Tổng hợp và nhóm dữ liệu
X Yêu cầu
Sử dụng DataFrame df từ tệp student.csv:
• Nhóm dữ liệu theo cột gender và tính trung bình của math.grade và tổng của
sciences.grade cho mỗi nhóm.
• Hiển thị kết quả dưới dạng DataFrame.
Kết quả là một DataFrame với các cột math.grade (trung bình) và sciences.grade (tổng),
được nhóm theo gender.

```python
df.groupby('gender').agg({'math.grade':'mean','sciences.grade':'sum'})
# nhóm dữ liệu theo cột gender và trung tính của math.grade và tổng của sciences.grade
```

Bài 3: Áp dụng hàm và lambda
X Yêu cầu
Sử dụng DataFrame df từ tệp student.csv:
• Tạo cột mới adjusted_age dựa trên cột age: nếu age > 18 thì tăng thêm 2, ngược lại
giảm đi 1.
• Tạo cột mới total_grade bằng tổng của english.grade, math.grade,
sciences.grade, và language.grade.
• Hiển thị 5 hàng đầu tiên của DataFrame sau khi thêm các cột mới.
Kết quả sẽ hiển thị DataFrame với các cột mới adjusted_age và total_grade.

```python
df['adjusted_age'] = df['age'].apply(lambda x: x+ 2 if x>18 else x+ 1)
df.describe()
```

```python
df['total_grade'] = df['english.grade']+df['math.grade']+df['sciences.grade']+df['language.grade']
df.head()
```

Sử dụng DataFrame df từ tệp student.csv.
• Tạo pivot table tính tổng math.grade theo nationality (hàng) và gender (cột).
• Tạo pivot table thứ hai tính trung bình english.grade theo ethnic.group (hàng) và
gender (cột).
• Hiển thị cả hai pivot table.
Kết quả: Hai pivot table với các giá trị tổng hợp tương ứng.

```python
df.pivot_table(
    values = 'math.grade',
    index = 'nationality',
    columns ='gender',
    aggfunc = 'sum'
)
```

```python
df.pivot_table(
    values = 'english.grade',
    index = 'ethnic.group',
    columns ='gender',
    aggfunc = 'mean'
)
```
