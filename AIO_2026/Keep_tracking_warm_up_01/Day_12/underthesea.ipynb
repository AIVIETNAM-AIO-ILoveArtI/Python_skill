```python
from underthesea import pos_tag
from underthesea import sent_tokenize
from underthesea import word_tokenize
from underthesea import classify
from underthesea import sentiment
# import thư viện cần dùng 
van_ban = 'Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thểtạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày.'
print(f"POS Tagging: {pos_tag(van_ban)}")
# pog_tag giúp dán nhãn từng từ
print(f"Text Classification: {classify(van_ban)}")
# classify giúp phân loại văn bản
print(f"Sentiment Analysis: {sentiment(van_ban)}")
# sentiment giúp phân tích cảm xúc
print(f"Sentence Segmentation: {sent_tokenize(van_ban)}")
# sent_tokenize giúp phân đoạn câu văn
print(f"Word Segmentation: {word_tokenize(van_ban)}")
# word_tokenize giúp phân đoạn từ ngữ
```
