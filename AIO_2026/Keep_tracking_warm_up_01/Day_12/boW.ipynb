Tạo Bag-Of-Word cho tập dataset sau: corpus = ["Tôi thích môn Toán", "Tôi
thích AI", "Tôi thích âm nhạc"]. Sau đó tạo list có tên vector để lưu vector sau
khi thực hiện bước Tokenization đoạn văn bản sau: Tôi thích AI thích Toán

```python
corpus = ["Tôi thích môn Toán", "Tôi thích AI", "Tôi thích âm nhạc"]
cau_noi = str()
# tạo sâu rỗng để lưu câu cuối cùng sau khi nối
for i in corpus:
    cau_noi += i + ' '
# tạo vòng lặp để nối câu
lst = list(set(cau_noi.split(' ')))
# tách các từ trong câu cuối cùng tổng kết được sao cho không có từ nào trùng nhau
bow_lst = [0]*len(lst)
# tạo vector lưu số lần xuất hiện của từng từ trong input
dauvao = input()
# nhập đầu vào
chia_tu = dauvao.split(' ')
# tạo list bởi các từ bị cắt từ input ra
for i in chia_tu:
    bow_lst[lst.index(i)]+=1
# tạo vòng lặp để đếm số lần xuất hiện từng từ trong đó
print(bow_lst)
# in kết quả
```
