Câu 1: Hãy viết chương trình tạo Numpy array, Tensorflow tensor, Pytorch tensor từ danh sách 1 chiều?

```python
import numpy as np
import torch
import tensorflow as tf
# import các thư viện cần dùng 
```

```python
lst = [1,2,3,4,5,6,7,8,9]
# tạo mảng 1 chiều trước để khởi tạo bằng các cách khác nhau
```

```python
np_1D_array = np.array(lst)
print(type(np_1D_array))
# xem danh sách 1 chiều vừa tạo thuộc kiểu nào
print(np_1D_array)
# in kết quả theo đề bài
```

```python
torch_1D_array = torch.tensor(lst)
print(type(torch_1D_array))
# xem tensor 1 chiều vừa tạo thuộc kiểu nào
print(torch_1D_array)
# in kết quả theo đề bài
```

```python
tensor_1D_array = tf.convert_to_tensor(lst)
print(type(tensor_1D_array))
# xem tensor 1 chiều vừa tạo thuộc kiểu nào 
print(tensor_1D_array)
# in kết quả theo đề bài
```

Câu 2: Hãy viết chương trình tạo Numpy array, Tensorflow tensor, Pytorch tensor từ danh sách 2 chiều. Sau đó thực hiện kiểm tra thuộc tính shape, dtype, type, device từ các array, tensor vừa tạo ?

```python
lst_2D = [[1,2,3],
          [4,5,6],
          [7,8,9]]
# tạo mảng 2 chiều trước để khởi tạo thành nhiều kiểu khác nhau
```

```python
np_2D_array = np.array(lst_2D)
print(f"kích thước của mảng np_2D_array là: {np_2D_array.shape}")
print(f"kiểu dữ liệu của mảng np_2D_array là: {np_2D_array.dtype}")
print(f"kiểu của mảng np_2D_array là: {type(np_2D_array)}")
```

```python
torch_2D_array = torch.tensor(lst_2D)
print(f"kích thước của torch_2D_array là: {torch_2D_array.shape}")
print(f"kiểu dữ liệu của torch_2D_array là: {torch_2D_array.dtype}")
print(f"kiểu của torch_2D_array là: {type(torch_2D_array)}")
print(f"thiết bị mà torch_2D_array lưu trữ: {torch_2D_array.device}")
```

```python
tensor_2D_array = tf.convert_to_tensor(lst_2D)
print(f"kích thước của tensor_2D_array là: {tensor_2D_array.shape}")
print(f"kiểu dữ liệu của tensor_2D_array là: {tensor_2D_array.dtype}")
print(f"kiểu của tensor_2D_array là: {type(tensor_2D_array)}")
print(f"thiết bị mà tensor_2D_array lưu trữ: {tensor_2D_array.device}")
```
