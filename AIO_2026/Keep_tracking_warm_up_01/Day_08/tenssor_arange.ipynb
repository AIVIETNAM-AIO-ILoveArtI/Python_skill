Câu 3: Hãy viết chương trình tạo Numpy array, Tensorflow tensor, Pytorch tensor ngẫu nhiên trong khoảng giá trị [0, 1] với kích thước (3, 4). Tiếp theo hãy tạo array, tensor với các giá trị là số nguyên trong khoảng [-10, 10]. Lưu ý trong bài tập này, các bạn sẽ sử dụng seed = 2024 để dảm bảo ra kết quả giống đáp án, cách sử dụng như sau:

```python
np.random.seed(2024)
print(np.random.rand(3,4))
print()
print(np.random.randint(-10, 10,size=(3,4)))
```

```python
torch.manual_seed(2024)
print(torch.rand(3,4))
print()
print(torch.randint(-10, 10,size=(3,4)))
```

```python
tf.random.set_seed(2024)
print(tf.random.uniform((3,4)))
print()
print(tf.random.uniform((3,4),minval=-10,maxval =10))
```
