```python
def find_none_index(data):
    result = -1
    # tạo biến lưu index xuất hiện giá trị None đầu tiên
    for i, v in enumerate(data):
        if v== None:
        # khi gặp giá trị None trong list data thì lưu vào biến
            result = i
            break
            # dừng vòng lặp ngay khi thấy để tránh cập nhật 

    return result

def find_end(data, index):
    end = len(data)
    for i in range(index+1, len(data), 1):
        if data[i] != None:
        # tìm giá trị không phải None có index gần với index đưa vào hàm nhất
            end = i
            break
    return end



def interpolate_nearest_neighbor(data, none_position):
    nearest_neighbor = None # tạo biến lưu giá trị lân cận 
    nearest_distance = float('inf') # tạo biến lưu khoảng cách 
    for index in range(len(data)):
    # duyệt để tìm giá trị lân cận gần nhất với giá trị có index
        if data[index] is not None and abs(index - none_position) < nearest_distance:
        # nếu thấy khoảng cách ngắn hơn với khoảng cách ngắn nhất thì thay thế
            nearest_neighbor = data[index] # thay giá trị lân cận
            nearest_distance = abs(index - none_position) # thay giá trị khoảng cách

    return nearest_neighbor

def process_data(data):
    while True:
        index = find_none_index(data)
        if index == -1:
            break
        # không tìm thấy giá trị None trả về -1

        # find begin
        begin = index - 1


        # find end
        end = find_end(data, index)

        if begin<0: # nếu giá trị None xuất hiện ở đầu
            y = data[end]

        elif end>=len(data): # nếu giá trị None xuất hiện ở cuối
            y = data[begin]
        else: # nếu giá trị None xuất hện ở giữa
            y = interpolate_nearest_neighbor(data, index)

        data[index] = y
    return data

# test
data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]
result = process_data(data)
print(result)
```
