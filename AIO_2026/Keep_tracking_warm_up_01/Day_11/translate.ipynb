```python
from translate import Translator as tsl1
from googletrans import Translator as tsl2
# import thư viện cần dùng
```

Hãy thực hiện dịch câu sau với 2 ngôn ngữ: Anh, Nhật sử dụng cả 2 thư viện: Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể tạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày.

* Dịch bằng thư viện translate

```python
dich = tsl1(to_lang='en')
dich = dich.translate('Anh, Nhật sử dụng cả 2 thư viện: Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể tạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày.')
print(f"bản dịch từ tiếng việt sang Tiếng Anh là: {dich}")
print()
dich2 = tsl1(to_lang='ja')
dich2 = dich2.translate('Anh, Nhật sử dụng cả 2 thư viện: Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể tạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày.')
print(f"bản dịch là từ tiếng việt sang Tiếng Nhật là: {dich2}")
```

* Dịch từ thư viện googletrans

```python
dich1 = tsl2()
dich1 = await dich1.translate('Anh, Nhật sử dụng cả 2 thư viện: Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể tạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày.',dest ='en')
print(f"bản dịch từ tiếng việt sang Tiếng Anh là: {dich1.text}")
print()
dich3 = tsl2()
dich3 = await dich3.translate('Anh, Nhật sử dụng cả 2 thư viện: Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể tạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày.',dest ='ja')
print(f"bản dịch từ tiếng việt sang Tiếng Anh là: {dich3.text}")
```
