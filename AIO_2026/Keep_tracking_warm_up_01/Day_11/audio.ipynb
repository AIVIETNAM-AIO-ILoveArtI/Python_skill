```python
lang = 'vi'
output_filename = 'record.mp3'
content = "Tôi yêu AIO"
output = gTTS(content,lang = lang,slow = False)
# dùng hàm gTTS để tạo file âm thanh
output.save(output_filename)
# lưu file âm thanh vào máy tính
data,sr2 = librosa.load(output_filename)
IPython.display.display( IPython.display.Audio(np.transpose(data),rate = sr2))
```

```python
audio,sr1 = librosa.load('record.mp3')
time = np . linspace (0 , audio.shape[0] / sr1 , num = audio.shape[0])
# dùng matplotlib để thể hiện biểu đồ âm thanh
plt . plot ( time , audio , color = 'blue')
plt . xlabel (" Thời gian (s)")
plt . ylabel (" Biên độ ( quantized )")
plt . title (" Biểu đồ âm thanh ")
plt . axhline ( y =0 , color = 'r', linestyle ='-')
plt . show ()
```
