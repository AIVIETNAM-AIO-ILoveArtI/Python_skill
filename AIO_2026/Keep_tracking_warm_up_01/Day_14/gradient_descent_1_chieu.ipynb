Bài tập 2: Lập trình

```python
import matplotlib.pyplot as plt
# import thư viện cần dùng
```

• Viết chương trình Python thực hiện GD cho f(x) = x^2

```python
def gradient_descent(xt,learning_rate):
    xt_next = xt - 2*xt*learning_rate
    return xt_next
```

• Khởi tạo x0 = 10, η = 0.1, và lặp 100 lần hoặc đến khi |Xt+1 − Xt| < 10^-6

```python
x0 = 10
learning_rate = 0.1
xt = []
# Tạo các list để lưu các xt
def loop(x0,learning_rate,xt):
    x1 = gradient_descent(x0,learning_rate)
    xt.append(x0)
    xt.append(x1)
    while abs(x1-x0)>=10**-6:
        x2 = gradient_descent(x1,learning_rate)
        x0 = x1
        x1 = x2
        xt.append(x2)
loop(x0,learning_rate,xt)
```

• In ra xt và f(Xt) sau mỗi 10 lần lặp.

```python
for i in range(0,len(xt),10):
    for j in xt[i:i+10]:
        print(j,end = ' ')
    print()
```

• Vẽ đồ thị f(x) = x^2 và đánh dấu các điểm xt theo các lần lặp (phần vẽ có thể tham khảo solution vì chưa học tới matplotlib).

```python
def f_x(x):
    f_x = x**2
    return f_x
f_xt = []
def f_lst(f_xt,xt):
    for i in range(len(xt)):
        f_xt.append(f_x(xt[i]))
f_lst(f_xt,xt)
# tính các f(xt) theo xt
def graph(xt,f_xt):
    plt.scatter(xt,f_xt)
    plt.xlabel('Xt')
    plt.ylabel('f(Xt)')
    plt.show()
graph(xt,f_xt)
# plot biểu đồ theo đề bài
```

Bài tập 3: Thử nghiệm với η

• Chạy lại chương trình với η = 0.01 và η = 0.5.

• So sánh tốc độ hội tụ (số lần lặp       để |xt| < 0.01) và rút ra kinh nghiệm chọn η trong quá trình training.

```python
learning_rate_1 = 0.01
learning_rate_2 = 0.5
# chạy với trường hợp tốc độ học là 0,01
xt_1 = []
f_xt_1 = []
loop(x0,learning_rate_1,xt_1)
f_lst(f_xt_1,xt_1)
graph(xt_1,f_xt_1)
# chạy với trường hợp tốc độ học là 0,5
xt_2 = []
f_xt_2 = []
loop(x0,learning_rate_2,xt_2)
f_lst(f_xt_2,xt_2)
graph(xt_2,f_xt_2)
```
