```python
lst_data = []
# Your code here
for i in range(2,13,2):
    lst_data.append(i)
    #thêm phần tử i ở cuối list data
print(lst_data)
for i in lst_data:
    if i%3==0:
        lst_data.remove(i)
        #xóa phần tử có giá trị i và có index thấp nhất
print(lst_data)
for i in range(1,4):
    lst_data.append(i)
    #thêm phần tử i vào cuối list data
for i in range(6,9):
    lst_data.insert(3,i)
    #thêm phần tử thứ i vào index=3 trong list data
print(lst_data)
for i in range(len(lst_data)):
#len(lst_data): độ dài list
    if lst_data[i]%2==0 or lst_data[i]%5==0:
        lst_data[i]=0
        #cập nhật 0 tại biến có index i
print(lst_data)
        
```
