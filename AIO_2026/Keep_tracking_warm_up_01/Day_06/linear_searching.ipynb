Tạo List có tên là lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]. Sau đó, áp dụng phương pháp tìm kiếm để tìm vị trí có giá trị None có trong lst_data theo 2 cách: tìm vị trí None đầu tiên, và tìm tất cả vị trí có giá trị None

```python
lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]
# tạo list lst_data theo đề bài
lst_index_of_None = []
# tạo list để lưu các chỉ số xuất hiện giá trị None
cnt = 0
# tạo biến đếm để tạo điều kiện in chỉ số đầu tiên xuất hiện
for i in range(len(lst_data)):
    if lst_data[i]==None:
        cnt+=1
        # cộng 1 lần biến đếm mỗi lần giá trị None xuất hiện
        lst_index_of_None.append(i)
        # lưu chỉ số index trong None
        if cnt==1: #khi xuất hiện vị trí đầu tiên có giá trị None thì in ra
            print("- Vị trí None đầu tiên:",i,end=' - ')
print(f"Danh sách vị trí có giá trị None: {lst_index_of_None}")
# in các index có giá trị None
```
