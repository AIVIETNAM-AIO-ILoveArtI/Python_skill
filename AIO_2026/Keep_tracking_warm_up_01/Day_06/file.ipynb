Câu 1: Hãy đọc file data1.txt có trong dataset phía trên và lưu vào biến data sau khi loại bỏ các ký tự \n và thay thế bằng khoảng trắng, đồng thời chuyển về chữ cái thường toàn bộ văn bản

```python
data = ''
# tạo một string rỗng để lưu đáp án
with open("data1.txt","r") as f:
# mở file data1.txt ở trạng thái đọc
    data1 = f.read()
    # lưu dữ liệu data1.txt vào biến data1
    data1 = data1.split("\n")
    # data1 biến thành list khi các câu được không còn tách bởi dấu xuống dòng
    for i in data1:
        data += i.lower()
        # mỗi từ được viết thường lại 
        data += ' '
    print(data)
    # in kết quả
```

Câu 2: Phân tích văn bản trên và lưu vào biến có tên distinct_words các chữ cái duy nhất trong câu

```python
distinct_words ={}
# tạo biến lưu các từ được xuất hiện riêng biệt trong biến data
data2 = data.split(' ')
# phân tách string bằng dấu cách dẫn đến data2 thành list
distinct_words = set(data2)
# chuyển list thành set
distinct_words = list(distinct_words)
# sau khi thành set có một phần tử 
distinct_words.sort()
# sort list theo string 
for i in distinct_words:
    print(i+' ',end ='')
# in kết quả
```

Câu 3: Đếm số lần từng chữ trong distinct_words xuất hiện trong văn bản. Và cho biết chữ nào xuất hiện nhiều nhất và ít nhất.

```python
cnt_word = []
# tạo list lưu chỉ từng chữ một
for i in range(len(distinct_words)):
    cnt_word.append(0)
# từng chữ xuất hiện trong danh sách đều có chỉ số = 0
cnt_each_word = dict(zip(distinct_words,cnt_word))    
# tạo dict để đếm các chữ cái riêng lẻ
for i in data2:
    cnt_each_word[i]+=1
    # đếm số lần xuất hiện của từng chữ
for key,value in cnt_each_word.items():
    print(f"+ \"{key}\" in data is {value}")
# in kết quả
```

```python
filter_data = list(cnt_each_word.items())
# biến dict thành list để sort
filter_data.sort(key = lambda x: (-x[1],x[0]))
# sort theo phần tử thứ 2 ở trên dòng từ lớn đến bé
print(f"+ \"{filter_data[0][0]}\" is the most frequent word")
# in phần tử đầu tức chữ cái xuất hiện nhiều nhất trong distinct_words
```

```python
print(f"+ \"{filter_data[-1][0]}\" is the least common word")
# in ra phần tử cuối cùng tức chữ cái xuất hiện ít nhất trong distinct_words
```
