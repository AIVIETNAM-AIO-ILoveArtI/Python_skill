```python
lst_data = [[1,2,3],
            [4,5,6],
            [7,8,9]]
#khởi tạo mảng 2 chiều 3x3
lst_sub_data =[]
#khởi tạo mảng 1 chiều
for i in range(0,3):
    lst_1 =[]
    #khởi tạo dòng nhằm thêm list_1 vào list_sub_data (mỗi list_1 là một dòng gồm 3 phần tử)
    #khi list trong list -> lst_sub_data thành mảng 2 chiều
    for j in range(0,3,2):
    #duyệt từ 0 tới 3 nhưng chỉ lấy phần tử mang index =0 và =2
    #->thêm bước đếm 2 ở cuối để nhảy thẳng ra vị trí có index = 2 ở mỗi dòng
        lst_1.append(lst_data[i][j])
        #thêm vào cuối lst_1 phần tử có vị trí ở i dòng và j cộy
    lst_sub_data.append(lst_1)
    #Thêm 1 dòng vào mảng 2 chiều
print(lst_sub_data)
#in kết quả
```

```python
mat_a = [[1,2,3],
         [4,5,6],
         [7,8,9]]
#khởi tạo ma trận A

mat_b = [[2,4,6],
         [1,3,5],
         [1,0,1]]
#khởi tạo ma trận B

mat_tong =[]
#khởi tạo mảng 1 chiều lưu đáp án tổng ma trận A và B

for i in range(0,3):
    mat_1chieu = []
    #Khởi tạo mảng 1 chiều tương đương với 1 dòng trong mảng 2 chiều ở trong mat_tong
    for j in range(0,3):
        mat_1chieu.append(mat_a[i][j]+mat_b[i][j])
        #Thêm vào mảng lưu kết quả ứng với vị trí từng phần tử trong một dòng i(mat_1chieu)
    mat_tong.append(mat_1chieu)
    #Thêm 1 dòng vào mảng 2 chiều

print(mat_tong)
#in kết quả tổng ma trận A+B

mat_hieu =[]
#khởi tạo mảng 1 chiều lưu kết quả hiệu ma trận A và B

for i in range(0,3):
    mat_1chieu = []
    #Khởi tạo mảng 1 chiều tương đương với 1 dòng trong mảng 2 chiều ở trong mat_hieu
    for j in range(0,3):
        mat_1chieu.append(mat_a[i][j]-mat_b[i][j])
         #Thêm vào mảng lưu từng kết quả ứng với vị trí từng phần tử trong dòng i(mat_1chieu)
    mat_hieu.append(mat_1chieu)
    #Thêm một dòng vào mảng 2 chiều

print(mat_hieu)
#in kết quả hiệu ma trận A - B

mat_tich =[]

for i in range(0,3):
    mat_1chieu =[]
    #Khởi tạo mảng 1 chiều tương đương với 1 dòng trong mảng 2 chiều ở trong mat_tich
    for j in range(0,3):
        ptu =0
        #Khởi tạo từng phần tử cho kết quả từng vị trí trong mat_tich
        for k in range(0,3):
        #Duyệt thêm vòng lặp k vì từng phần tử được tính bằng tính bằng tổng từng phần tử dòng A nhân từng phần tử dòng B
        #Vòng lặp k sinh ra nhằm tính toán từng phần tử cho từng vị trí
            ptu += mat_a[i][k]*mat_b[k][j]
            #cộng dồn để được ptu cuối cùng
        mat_1chieu.append(ptu)
        #sau khi vòng lặp k chạy xong thì ta có kết quả ptu ở vị trí mat_tich[i][j]
    mat_tich.append(mat_1chieu)
    #Thêm dòng vào mảng 2 chiều

print(mat_tich)
#in kết quả tích ma trận A cho B (A,B không có tính giao hoán)
            
```
