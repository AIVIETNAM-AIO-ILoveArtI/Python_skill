```python
import random
def random_number_with_condition ( total ) :
    # Set a random value that is the same between devices
    random . seed (0)
    # random.seed(0)để cố định kết quả ra khi mà hàm random trả về kết quả ngẫu nhiên
    a = random.randint(1,20)
    b = random.randint(1,20)
    # chọn a,b ngẫu nhiên từ khoảng 1 tới 20
    cntloop =1
    # tạo biến đếm vòng lặp a,b thay đổi
    while a+b!=total:
        a = random.randint(1,20)
        b = random.randint(1,20)
        cntloop +=1
    return cntloop
    #trả về kết quả
print(random_number_with_condition(40))
print(random_number_with_condition(35))
print(random_number_with_condition(20))
#in kết quả
```

```python
def find_divisible_number ( a) :
    """
    Find the smallest integer that is greater than 100 and divisible
    by the integer a
    """

    # Your code here
    i = 101
    #khởi tạo biến lớn hơn 100 để tìm số lớn hơn 100
    while i%a!=0:
        i+=1
    #duyệt từ 100 đổ lên cho tới khi nào 
    return i
    #trả về kết quả
print(find_divisible_number(5))
print(find_divisible_number(17))
#in kết quả
```

```python
def find_squared_root ( a ) :
    """ Find the squared root of number a """
    EPSILON = 0.001

    #Your code is here
    xn = a
    xn1 = xn - (xn**2-a)/(2*xn)
    #dao ham cua ham f(x)=x^2 -a là 2*x 
    while abs(xn1-xn)>=EPSILON:
    #khi nào căn chưa tối ưu thì chưa dừng
        xn = xn1
        xn1 = xn - (xn**2-a)/(2*xn)
        
    return xn1
    #trả về kết quả
print(find_squared_root(2))
print(find_squared_root(3))
#in kết quả
    



```
