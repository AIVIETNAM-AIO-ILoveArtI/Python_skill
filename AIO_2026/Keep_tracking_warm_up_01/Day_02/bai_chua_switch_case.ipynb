```python
time = input()
#nhập input
if time =='5 AM':
#nếu là 5 giờ sáng
    print("Wake up")
elif time =='6 AM':
#nếu không phải là 5 AM THÌ MỚI CHẠY
    print("Yoga")
elif time =='7 AM':
#nếu không phải là 6 AM THÌ MỚI CHẠY    
    print("Work")
else:
#nếu không phải các trường hợp trên
    print("Do something else")

```

```python
todo ={"5 AM":"Wake up",
       "6 AM":"Yoga",                 
       "7 AM":"Work"}
#tạo dictionary nhằm lưu mỗi thời gian tương ứng với mỗi hoạt động
time = input()
#nhập thời gian
todo.get(time,"Do something else")
#hàm get giúp tìm thời gian tương ứng trong dict "todo" nếu không có thì sẽ in ra thứ sau dấu phẩy
```

```python
time = input()
match time:
    case "5 AM":
    #với trường hợp là 5 AM
        print("Wake up")
    case "6 AM":
    #với trường hợp là 6 AM
        print("Yoga")
    case "7 AM":
    #với trường hợp là 7 AM
        print("Work")
    case _:
    #với trường hợp khác các trường hợp trên
        print("Do something else")
```
