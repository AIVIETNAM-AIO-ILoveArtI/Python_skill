```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
# import thư viện
```

Hãy đọc và hiển thị ảnh có tên 2.jpg trong tập dữ liệu trên, áp dụng kỹ thuật Gaussian Filter trong thư viện OpenCV theo 2 cách cài đặt với các sigma khác nhau như [2.5, 5.0, 10.0]

```python
# code theo hàm cho sẵn
def gaussian_kernel (size,sigma) :
    if size % 2 == 0:
        size = size + 1
    max_point = size // 2 # both directions (x,y) maximum cell start point
    min_point = - max_point # both directions (x,y) minimum cell start point
    K = np.zeros((size,size)) # kernel matrix
    for x in range ( min_point , max_point + 1) :
        for y in range ( min_point , max_point + 1) :
            # ghi biểu thức cần điền vào
            value = (1 / (2 * np.pi * sigma**2)) * np.exp(-(x**2 + y**2) / (2 * sigma**2))
            K [ x - min_point , y - min_point ] = value
    
    return K

kernel = gaussian_kernel(5,1.4)
img = cv2 . imread (r"C:\Users\ldraw\Documents\AIO 2026\Keep_tracking_warm_up_1\Day 10\dataset\2.jpg", 0)
img_gaussian = cv2.filter2D(img,-1,kernel )
plt.imshow(img_gaussian,cmap='gray')
plt.show()
```

```python
img_filtered = cv2.GaussianBlur(img,(61 ,61),2.5)
plt.imshow(img_gaussian,cmap='gray')
plt.show()
```
